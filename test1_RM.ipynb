```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Downloads/data.csv')
df
```

```python
df['LOCATION'].unique()
```

```python
ex_returns = df.pivot(index='TIME_M', columns='LOCATION', values='FX')\
            .reset_index()
ex_returns.columns.name=None
ex_returns.set_index('TIME_M')
ex_returns
```

```python
returns = df.pivot(index='TIME_M', columns='LOCATION', values='CPI')\
            .reset_index()
returns.columns.name=None
returns.set_index('TIME_M')
returns=returns.dropna()
returns
```

```python
r_df=returns['TIME_M']
new_ex=pd.merge(r_df, ex_returns, on=['TIME_M'], how='left')
new_ex
```

```python
colm=['FX_AUS', 'FX_AUT', 'FX_BEL', 'FX_CAN', 'FX_CHL', 'FX_CZE',
       'FX_DNK', 'FX_EA19', 'FX_FIN', 'FX_FRA', 'FX_DEU', 'FX_GRC',
       'FX_HUN', 'FX_IND', 'FX_IDN', 'FX_IRL', 'FX_ITA', 'FX_JPN',
       'FX_KWT', 'FX_MAL', 'FX_MEX', 'FX_NLD', 'FX_NZL', 'FX_NOR',
       'FX_PHP', 'FX_POL', 'FX_PRT', 'FX_SAU', 'FX_SNG', 'FX_ZAR',
       'FX_KOR', 'FX_ESP', 'FX_SWE', 'FX_CHE', 'FX_TWN', 'FX_THA',
       'FX_TUR', 'FX_UAE', 'FX_GBR']
```

```python
p01=np.quantile(returns[colm], 0.1, axis=1)
p02=np.quantile(returns[colm], 0.2, axis=1)
p04=np.quantile(returns[colm], 0.4, axis=1)
p06=np.quantile(returns[colm], 0.6, axis=1)
p08=np.quantile(returns[colm], 0.8, axis=1)
p09=np.quantile(returns[colm], 0.9, axis=1)
```

```python
#group values in 5 portfolios
rows_p1=[]
rows_p2=[]
rows_p3=[]
rows_p4=[]
rows_p5=[]
rows_p1r=[]
rows_p5r=[]
#for col in returns[colm].columns:
for i in range(0, len(p02)):
    rows_p1.append(returns[returns[colm]<=p02[i]].iloc[i].fillna(0))
    rows_p2.append(returns[(returns[colm]>=p02[i]) & (returns[colm] <= p04[i])].iloc[i].fillna(0))
    rows_p3.append(returns[(returns[colm]>=p04[i]) & (returns[colm] <= p06[i])].iloc[i].fillna(0))
    rows_p4.append(returns[(returns[colm]>=p06[i]) & (returns[colm] <= p08[i])].iloc[i].fillna(0))
    rows_p5.append(returns[returns[colm]>=p08[i]].iloc[i].fillna(0))
    rows_p1r.append(returns[returns[colm]<=p01[i]].iloc[i].fillna(0))
    rows_p5r.append(returns[returns[colm]>=p09[i]].iloc[i].fillna(0))
```

```python
#creating df
df_p1=pd.DataFrame(rows_p1, columns=colm)
df_p2=pd.DataFrame(rows_p2, columns=colm)
df_p3=pd.DataFrame(rows_p3, columns=colm)
df_p4=pd.DataFrame(rows_p4, columns=colm)
df_p5=pd.DataFrame(rows_p5, columns=colm)
df_p1r=pd.DataFrame(rows_p1r, columns=colm)
df_p5r=pd.DataFrame(rows_p5r, columns=colm)
```

```python
#creating dummy df to extract returna and calculate the portfolios returns
dum_p1 = df_p1.where(df_p1[colm] == 0, 1)
dum_p5 = df_p5.where(df_p5[colm] == 0, 1)
dum_p1r = df_p1r.where(df_p1r[colm] == 0, 1)
dum_p5r = df_p5r.where(df_p5r[colm] == 0, 1)
dum_p5
```

```python
#creating portfolios and calculating portfolios` returns 1/n*a+1/n*b=1/n*(a+b)
p1=dum_p1[colm].mul(new_ex[colm].values)
p5=dum_p5[colm].mul(new_ex[colm].values)
p1r=dum_p1r[colm].mul(new_ex[colm].values)
p5r=dum_p5r[colm].mul(new_ex[colm].values)
p1['returns_mean_p1']=p1.sum(axis=1)/p1[p1[colm] != 0].count(axis=1)
p1['rob_returns_mean_p1']=p1r.sum(axis=1)/p1r[p1r[colm] != 0].count(axis=1)
p5['returns_mean_p5']=p5.sum(axis=1)/p5[p5[colm] != 0].count(axis=1)
p5['rob_returns_mean_p5']=p5r.sum(axis=1)/p5r[p5r[colm] != 0].count(axis=1)
p1['TIME_M']=returns['TIME_M'].values
p5['TIME_M']=returns['TIME_M'].values
```

```python
d_p1=p1[['TIME_M','returns_mean_p1','rob_returns_mean_p1']]
d_p5=p5[['TIME_M','returns_mean_p5','rob_returns_mean_p5']]
diff=pd.merge(d_p1, d_p5, on='TIME_M', how='outer')
diff['difference_usual']=diff['returns_mean_p1']-diff['returns_mean_p5']
diff['difference_rob']=diff['rob_returns_mean_p1']-diff['rob_returns_mean_p5']
diff
```

```python
q=np.cumsum(diff['returns_mean_p1'])
p=np.cumsum(diff['returns_mean_p5'])
plt.plot(diff['TIME_M'], q)
plt.plot(diff['TIME_M'], p)
plt.ylabel('Exchange_Rate_returns')
plt.xlabel('time')
plt.legend(['y = returns_mean_p1', 'y = returns_mean_p5'],  loc='upper left')
plt.show()
```
